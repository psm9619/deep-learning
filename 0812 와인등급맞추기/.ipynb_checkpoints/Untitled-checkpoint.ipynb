```python
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from keras.utils import np_utils
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tensorflow as tf
import os
from sklearn.model_selection import train_test_split
```

```python
seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)
```

```python
df_pre = pd.read_csv("../data/wine.csv", header = None)
df_pre.head()
```

```python
df = df_pre.sample (frac=0.30)

# 와인등급은 11번째 칼럼이므로 X에 11번째를 제외한 나머지를 넣는 과정이 까다롭다.
dataset_1 = df.values
dataset_2 = df.values
X = np.delete(dataset_1, 11, axis=1)
Y = dataset_2[:,11]
X[1]
```

```python
Y_encoded = np_utils.to_categorical(Y)
Y_encoded[1]
```

```python
model = Sequential()
model.add(Dense(60, input_dim = 12, activation = "relu"))
model.add(Dense(24, activation = "relu"))
# model.add(Dense(12, activation = "relu"))
model.add(Dense(10, activation = "softmax"))

model.compile(loss = "categorical_crossentropy",
             optimizer = "adam",
             metrics = ["accuracy"])
```

```python
# MODEL_DIR = "./model1/"
# if not os.path.exists(MODEL_DIR):
  #  os.mkdir(MODEL_DIR)
    
# modelpath = "./model1/{epoch:02d}-{val_loss:4f}.hdf5"
```

```python
early_stopping_callback = EarlyStopping (monitor = "val_loss", verbose=0, patience = 100)

# checkpointer = ModelCheckpoint(filepath = modelpath, monitor = "val_loss", verbose = 1, save_best_only = True)
```

```python
history = model.fit(X, Y_encoded, validation_split = 0.2, epochs=2000, batch_size=30, 
                    callbacks = [early_stopping_callback])
```

```python
model.evaluate(X,Y_encoded)[1]
```

```python
y_vloss = history.history['val_loss']
y_acc = history.history['acc']
```

```python
x_len = np.arange(len(y_acc))
plt.plot(x_len, y_vloss, "o", c="red", markersize = 3)
plt.plot(x_len, y_acc, "o", c="blue", markersize = 3)
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_encoded, test_size=0.3, random_state=seed)
```

```python
model.fit(X_train, Y_train, epochs = 2000, batch_size=30)
```

```python
model.save("my_model.h5")
```
