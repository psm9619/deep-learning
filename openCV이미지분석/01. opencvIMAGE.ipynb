```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("../data/images/cat.jpg", cv2.IMREAD_COLOR)
plt.imshow(image)
```

```python
cat = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(cat)
```

```python
b,g,r = cv2.split(image)
image2 = cv2.merge([r,g, b])
plt.imshow(image2)
```

```python
# Gray scale image
grayimage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(grayimage,cmap=plt.get_cmap('gray'))
```

### 특정 범위 픽셀 변경

```python
image = cv2.imread("../data/images/cat.jpg", cv2.IMREAD_COLOR)
import time
start_time = time.time()
for i in range(0,100):
    for j in range(0,100):
        image[i,j] = [255,255,255]
# 특정범위 픽셀변경에 걸리는 시간 계산
print("---%s seconds---"%(time.time()-start_time))
```

```python
basicimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(basicimage)
```

```python
# make a copy before another action
basicImage= basicimage 
start_time = time.time()
basicImage[0:100, 0:100] = [0, 0, 0]
print("--- %s seconds ---" % (time.time() - start_time))

plt.imshow(basicImage)
```
